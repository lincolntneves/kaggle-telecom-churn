```python
X_train = train.set_index('id')[feature_names]
X_train.head()
```

```python
X_train = train.set_index('id')[feature_names].fillna(train.median())
X_train.head()
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X_train)
```

```python
y = train['churn_probability']
```

```python
%%time
clf_LR = LogisticRegression(random_state=12345)
clf_LR.fit(X,y)
```

## Predict test dataset with the first ML model

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
test_ = test.set_index('id')[feature_names].fillna(test.median())
test_.head()
```

```python
X_test = scaler.fit_transform(test_)
```

```python
preds_LR = clf_LR.predict(X_test)
```

```python
preds_LR_ = pd.DataFrame(preds_LR, columns=['churn_probability'])
```

```python
sub21_LR = pd.concat([test['id'],preds_LR_], axis=1)
```
