```python
import pandas as pd
import os
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_predict
```

```python
## Dimensões para visualização de datasets
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.float_format', '{:.8f}'.format)
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
train.info(verbose=True)
```

```python
train.churn_probability.value_counts()
```

```python
7132/(7132+62867)
```

```python
# Dropping dataframe columns base on dtype = object
train_ = train.select_dtypes(exclude=['object'])
train_.info(verbose=True)
```

```python
train_.shape
```

```python
train_.isnull().sum()
```

```python
# Replace all missing values with average of columns
train_ = train_.fillna(train_.median())
train_.isnull().sum()
```

```python
train_ = train_.set_index('id')
train_.head()
```

```python
X_train = train_.iloc[:,:-1]
X_train.head()
```

```python
y = train_['churn_probability']
y
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X_train)
```

```python
X
```

```python
clf = RandomForestClassifier(random_state=12345)
clf.fit(X,y)
```

```python
pd.DataFrame(zip(X_train.columns, clf.feature_importances_),columns=['Variavel', 'feature_importances']).sort_values('feature_importances', ascending=False)
```

```python
resultados = cross_val_predict(clf, X, y, cv = 5)
print(metrics.accuracy_score(y, resultados))
```

```python
print(metrics.classification_report(y, resultados))
```
