```python
import pandas as pd
import os
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_predict
from sklearn.linear_model import LogisticRegression
```

```python
## Dimensões para visualização de datasets
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.float_format', '{:.8f}'.format)
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
train.info(verbose=True)
```

```python
feature_date_in = [
    'date_of_last_rech_6', 'date_of_last_rech_7', 'date_of_last_rech_8',
    'date_of_last_rech_data_6','date_of_last_rech_data_7','date_of_last_rech_data_8'
]
```

```python
train[feature_date_in] = train[feature_date_in].apply(pd.to_datetime)
train.head()
```

```python
train[feature_date_in].info()
```

```python
train['date_of_last_rech_6_d'] = train['date_of_last_rech_6'].dt.day
train['date_of_last_rech_7_d'] = train['date_of_last_rech_7'].dt.day
train['date_of_last_rech_8_d'] = train['date_of_last_rech_8'].dt.day

train['date_of_last_rech_6_d_rep'] = train['date_of_last_rech_6_d'] / 30
train['date_of_last_rech_7_d_rep'] = train['date_of_last_rech_7_d'] / 31
train['date_of_last_rech_8_d_rep'] = train['date_of_last_rech_8_d'] / 31

train['date_of_last_rech_data_6_d'] = train['date_of_last_rech_data_6'].dt.day
train['date_of_last_rech_data_7_d'] = train['date_of_last_rech_data_7'].dt.day
train['date_of_last_rech_data_8_d'] = train['date_of_last_rech_data_8'].dt.day

train['date_of_last_rech_data_6_d_rep'] = train['date_of_last_rech_data_6_d'] / 30
train['date_of_last_rech_data_7_d_rep'] = train['date_of_last_rech_data_7_d'] / 31
train['date_of_last_rech_data_8_d_rep'] = train['date_of_last_rech_data_8_d'] / 31
    
train.head()
```

```python
features_names = [
"total_ic_mou_8",
"total_og_mou_8",
"arpu_8",
"roam_ic_mou_8",
"total_rech_amt_8",
"max_rech_amt_8",
"roam_og_mou_8",
"og_others_8",
"loc_ic_t2m_mou_8",
"loc_ic_mou_8",
"isd_og_mou_8",
"total_rech_num_8",
"isd_ic_mou_8",
"last_day_rch_amt_8",
"loc_ic_t2t_mou_8",
"onnet_mou_8",
"loc_og_mou_8",
"loc_og_t2m_mou_8",
"loc_og_t2t_mou_8",
"offnet_mou_8",
"std_og_t2m_mou_8",
"std_og_t2f_mou_8",
"loc_og_t2c_mou_8",
"std_ic_mou_8",
"std_og_mou_8",
"std_og_t2t_mou_8",
"std_ic_t2t_mou_8",
"vol_2g_mb_8",
"std_ic_t2m_mou_8",
"fb_user_8",
"spl_og_mou_8",
"loc_ic_t2f_mou_8",
"loc_og_t2f_mou_8",
"ic_others_8",
"std_ic_t2f_mou_8",
"spl_ic_mou_8",
"vol_3g_mb_8",
"max_rech_data_8",
"night_pck_user_8",
"count_rech_2g_8",
"av_rech_amt_data_8",
"arpu_2g_8",
"total_rech_data_8",
"arpu_3g_8",
"count_rech_3g_8",
"sachet_2g_8",
"monthly_3g_8",
"sachet_3g_8",
"monthly_2g_8",
'date_of_last_rech_6_d_rep',
'date_of_last_rech_7_d_rep',
'date_of_last_rech_8_d_rep',
'date_of_last_rech_data_6_d_rep',
'date_of_last_rech_data_7_d_rep',
'date_of_last_rech_data_8_d_rep'
    
]
```

```python
X_train = train.set_index('id')[features_names].fillna(train.mean())
X_train.head()
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X_train)
```

```python
X
```

```python
y = train['churn_probability']
y
```

```python
clf = LogisticRegression()
clf.fit(X,y)
```

```python
resultados = cross_val_predict(clf, X, y, cv = 5)
print(metrics.accuracy_score(y, resultados))
```

```python
print(metrics.classification_report(y, resultados))
```
