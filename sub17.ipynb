```python
X_train1 = train.set_index('id')[features_names_LessMissing].fillna(train.mean())
X_train1.head()
```

```python
X_train2 = train.set_index('id')[features_names_BigMissing].fillna(train.median())
X_train2.head()
```

```python
X_train = pd.concat([X_train1,X_train2], axis=1)
X_train.head()
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X_train)
```

```python
y = train['churn_probability']
```

```python
%%time
clf_RF = RandomForestClassifier(random_state=12345)
clf_RF.fit(X,y)
```

```python
%%time
clf_LR = LogisticRegression(random_state=12345)
clf_LR.fit(X,y)
```
